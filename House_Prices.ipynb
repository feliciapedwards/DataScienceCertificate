TASK: We are given the following situation: imagine if house pricing was as easy as a house costs 50k + 50k per bedroom, so that a 1 bedroom house costs 100k, a 2 bedroom house costs 150k etc. We want to figure out how much a 7 bedroom house would cost.

This is a house model below. 
Here I use the "Dense" to define a of connected neurons. I've only defined one dense, so there's only one layer, and only one unit, so it's a single neuron. I've compiled the neural network using the loss function 'means_squared_error', and the optimizer 'sgd'. These functions will help the neural network figure out the pattern and measure the "goodness" of the guess, before guessing again on the next epoch.

```
def house_model(y_new):
    xs = np.array([-1.0,0.0,1.0,2.0,3.0], dtype=float)
    ys = np.array([0.0,0.50,1.00,1.50,2.00], dtype=float)
    model =  tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(xs, ys, epochs=500)
    return model.predict(y_new)[0]
```

I want to predict the price of a house that has 7 (seven) rooms.

```
prediction = house_model([7.0])
print(prediction*100000)
```

Looks like a 7 bedroom house would cost around $400,281. This is pretty cheap if it's in Seattle area. I would like to know where I could buy a house so cheap!
